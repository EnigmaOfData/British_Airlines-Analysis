```python
# Set the style for the plots
sns.set_style("whitegrid")

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create a count plot for the recommended column, grouped by seat_type
sns.countplot(x='seat_type', hue='recommended', data=df, ax=ax)

# Set the title and labels
ax.set_title('Number of Recommendations by Seat Type')
ax.set_xlabel('Seat Type')
ax.set_ylabel('Number of Recommendations')

# Compute the total number of customers for each seat type
total_customers = df['seat_type'].value_counts()

# Add the total number of customers above the bars
for i, p in enumerate(ax.patches):
    x = p.get_x() + p.get_width() / 2
    y = p.get_y() + p.get_height()
    label = total_customers.iloc[i // 2]
    ax.text(x, y, label, ha='center', va='bottom')

# Show the plot
plt.show()
```

Shows how different seat type people have recomended the airlines

```python
# Set the style for the plots
sns.set_style("whitegrid")

# Compute the average of all service ratings for each seat type
avg_ratings = df.groupby('seat_type')[['seat_comfort', 'cabin_staff_service', 'food_beverages', 'inflight_entertainment', 'value_for_money']].mean().mean(axis=1)

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar plot for the average of all service ratings by seat type
sns.barplot(x=avg_ratings.index, y=avg_ratings.values, ax=ax)

# Set the title and labels
ax.set_title('Average of All Service Ratings by Seat Type')
ax.set_xlabel('Seat Type')
ax.set_ylabel('Average of All Service Ratings')

# Show the plot
plt.show()
```

Solo Leisure customers seems to have given a good rating to overall services

```python
from textblob import TextBlob

# Define a function to classify a review as positive or negative
def classify_review(review):
    # Compute the sentiment polarity score for the review
    polarity = TextBlob(review).sentiment.polarity

    # Classify the review as positive or negative based on the polarity score
    if polarity > 0:
        return 'positive'
    else:
        return 'negative'

# Apply the classify_review function to the reviews column
df['sentiment'] = df['reviews'].apply(classify_review)
```

```python
display(df.head())
```

```python
# Set the style for the plots
sns.set_style("whitegrid")

# Compute the proportion of positive and negative reviews for each seat type
positive_proportion = df[df['sentiment'] == 'positive'].groupby('seat_type')['sentiment'].count() / df.groupby('seat_type')['sentiment'].count()
negative_proportion = df[df['sentiment'] == 'negative'].groupby('seat_type')['sentiment'].count() / df.groupby('seat_type')['sentiment'].count()

# Create a figure and axes
fig, ax = plt.subplots(figsize=(10, 6))

# Create a bar plot for the proportion of positive and negative reviews by seat type
sns.barplot(x=positive_proportion.index, y=positive_proportion.values, color='blue', label='Positive', ax=ax)
sns.barplot(x=negative_proportion.index, y=negative_proportion.values, color='red', label='Negative', ax=ax)

# Set the title and labels
ax.set_title('Proportion of Positive and Negative Reviews by Seat Type')
ax.set_xlabel('Seat Type')
ax.set_ylabel('Proportion of Reviews')
ax.legend()

# Show the plot
plt.show()

```

Solo Leisure customer seems to have written most positive reviews which is not contradicting the rating given by them to overall services
